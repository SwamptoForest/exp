```python
import plotly.graph_objects as go
```

```python
fig = go.Figure() # 빈 도화지
```

```python
fig.show()
```

```python
fig.add_trace(go.Scatter(x = [1, 2, 3, 4], y = [10, 15, 20, 25], mode = 'lines+markers', name = '그래프그래프')) #Scatter - 선그래프
```

```python
fig.add_trace(go.Bar(x=[1,2,3,4], y=[13,16,34,99], name = '막대')) # Bar - 막대그래프
```

```python
fig.show()
```

```python
fig.update_traces(opacity=0.7)
fig.show()
```

```python
fig.update_traces(hovertemplate='얼마요: %{y}')

fig.show()
```

```python
fig.update_traces(marker_color='yellow', selector=dict(type='scatter'))

fig.show()
```

```python
fig.update_traces(line_width=8)
```

```python
fig.update_traces(line_width=8, selector=dict(type='line'))
```

```python
fig.update_traces(line_width=4, selector={'type': 'scatter'})
```

```python
# 축 지정하고 삭제하고 수정해보기
fig.update_yaxes(visible = False)
fig.show()
```

```python
fig.update_yaxes(visible = False)
fig.show()
```

```python
fig.update_xaxes(
    range=[0, 20],            # [범위 지정] 0부터 6까지만 보여줘 (줌인 효과)
    title_text="가 로 제 목",  # [수정] 축 제목 달기
    showgrid=True,           # [수정] 격자 무늬 켜기
    gridwidth=2,             # [수정] 격자 두께
    gridcolor='purple',   # [수정] 격자 색상
    zeroline=True,           # [수정] 0점 기준선 표시
    zerolinecolor='red'      # [수정] 0점 기준선 빨간색으로
)
```

```python
# 프로틀리프롯리로 안해본 것 해보기.
# 일단 제미나이로 예시 만듬


fig = go.Figure(go.Sunburst(
    labels=["지구", "아시아", "유럽", "한국", "일본", "프랑스", "독일", "서울", "부산", "도쿄", "파리", "베를린"],
    parents=["", "지구", "지구", "아시아", "아시아", "유럽", "유럽", "한국", "한국", "일본", "프랑스", "독일"],
    values=[0, 0, 0, 0, 0, 0, 0, 950, 330, 1400, 210, 360],
    
    # [수정] 이 옵션을 지우면 Plotly가 알아서 0인 부모 값을 자식들의 합으로 계산해 줌!
    # branchvalues="total"  <-- 이 줄 삭제함
))

fig.update_layout(title="계층형 데이터 시각화 (Sunburst) - 수정버전")
fig.show()
```

```python
# 어디에 쓴다는건지 설명을 봐도 확 알겠다는 느낌은 안든다. 그래도 따라해보기
fig = go.Figure(go.Sunburst(
    labels=['태양','수성','금성','지구','화성','목성','토성','천왕성','해왕성','명왕성'],
    parents=['','태양','수성','금성','지구','화성','목성','토성','천왕성','해왕성','명왕성'],))

fig.update_layout(title= '태양계는 도넛이다')
fig.show()
```
