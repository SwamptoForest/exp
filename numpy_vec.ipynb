```python
import numpy as np  # ë„˜íŒŒì´ë¥¼ ë¶ˆëŸ¬ì˜¤ê³  'np'ë¼ëŠ” ë³„ëª…ìœ¼ë¡œ ë¶€ë¥´ê¸°ë¡œ ì•½ì†

# 1. ê·¸ëƒ¥ íŒŒì´ì¬ ë¦¬ìŠ¤íŠ¸ë¡œ ë”í•˜ê¸°ë¥¼ í•´ë³´ì (ìš°ë¦¬ê°€ ì›í•œ ê²Œ ì•„ë‹˜)
list_a = [10, 20, 30]
list_b = [1, 2, 3]
print(f"íŒŒì´ì¬ ë¦¬ìŠ¤íŠ¸ ë”í•˜ê¸°: {list_a + list_b}") 
# ê²°ê³¼: [10, 20, 30, 1, 2, 3] -> ê·¸ëƒ¥ ì´ì–´ ë¶™ì—¬ë²„ë¦¼... ë©ì²­ì´.

print("-" * 30)

# 2. ë„˜íŒŒì´ë¡œ ë²¡í„°(ë°°ì—´) ë§Œë“¤ê¸°
# np.array() : "ì•¼, ì´ ë¦¬ìŠ¤íŠ¸ë¥¼ ê³„ì‚° ê°€ëŠ¥í•œ 'ë„˜íŒŒì´ ë°°ì—´'ë¡œ ë³€ì‹ ì‹œì¼œ!"
vec_a = np.array([10, 20, 30])
vec_b = np.array([1, 2, 3])

print(f"ë²¡í„° A: {vec_a}")
print(f"ë²¡í„° B: {vec_b}")

print("-" * 30)

# 3. ë²¡í„° ì—°ì‚° ìˆ˜í–‰ (ì´ê²Œ ìš°ë¦¬ê°€ ì›í•˜ë˜ ìˆ˜í•™ì´ì§€!)
# ê°™ì€ ìë¦¬(ì¸ë±ìŠ¤)ì— ìˆëŠ” ìˆ«ìë¼ë¦¬ ê³„ì‚°ë¨

# ë§ì…ˆ
add_result = vec_a + vec_b
print(f"ë§ì…ˆ ê²°ê³¼ (A + B): {add_result}") 
# 10+1, 20+2, 30+3 -> [11, 22, 33]

# ëº„ì…ˆ
sub_result = vec_a - vec_b
print(f"ëº„ì…ˆ ê²°ê³¼ (A - B): {sub_result}")
# 10-1, 20-2, 30-3 -> [9, 18, 27]

# ê³±ì…ˆ
mul_result = vec_a * vec_b
print(f"ê³±ì…ˆ ê²°ê³¼ (A * B): {mul_result}")
# 10*1, 20*2, 30*3 -> [10, 40, 90]
```

```python
import numpy as np

# ê°œìˆ˜ê°€ ë‹¤ë¥¸ ë‘ ë²¡í„° (2ê°œ vs 3ê°œ)
vec_short = np.array([10, 20])       # (x, y)
vec_long  = np.array([1, 2, 3])      # (x, y, z)

print("--- ê³„ì‚° ì‹œë„ ---")
# ì—¬ê¸°ì„œ ì—ëŸ¬ ë°œìƒ!
# ValueError: operands could not be broadcast together with shapes (2,) (3,)
try:
    result = vec_short + vec_long
    print(result)
except ValueError as e:
    print("ğŸš¨ ì—ëŸ¬ ë°œìƒ! ğŸš¨")
    print(f"ì—ëŸ¬ ë©”ì‹œì§€: {e}")
```

```python
# 1. ê³„ìˆ˜ í–‰ë ¬ (x, y ì•ì˜ ìˆ«ìë“¤)
# [1,  1]
# [1, -1]
A = np.array([
    [1, 1], 
    [1, -1]
])

# 2. ê²°ê³¼ ë²¡í„° (ëŠ”(=) ë’¤ì˜ ìˆ«ìë“¤)
b = np.array([5, 1])

# 3. í’€ì–´ë¼! (solve)
solution = np.linalg.solve(A, b)

print(solution) 
# ê²°ê³¼: [3. 2.]  <- x=3, y=2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Xì¶•ì˜ ë²”ìœ„ ì„¤ì • (0ë¶€í„° 6ê¹Œì§€)
x = np.linspace(0, 6, 100)

# 2. ìš°ë¦¬ì˜ ì‹ì„ y = ... í˜•íƒœë¡œ ë°”ê¾¼ ê²ƒ
# ì‹1: x + y = 5  ->  y = -x + 5
# ì‹2: x - y = 1  ->  y = x - 1
y1 = -x + 5
y2 = x - 1

# 3. ê·¸ë¦¼ ê·¸ë¦¬ê¸° ì‹œì‘
plt.figure(figsize=(6, 6)) # ë„í™”ì§€ í¬ê¸° ì„¤ì •

# ì„  ê¸‹ê¸°
plt.plot(x, y1, label='x + y = 5 (hap = 5)', color='blue')   # íŒŒë€ì„ 
plt.plot(x, y2, label='x - y = 1 (cha = 1)', color='orange') # ì£¼í™©ì„ 

# 4. ì •ë‹µ ìœ„ì¹˜ì— ì  ì°ê¸° (x=3, y=2)
plt.plot(3, 2, 'ro', markersize=10, label='Answer (3, 2)') # ë¹¨ê°„ ì (ro)

# ê¾¸ë¯¸ê¸° (ê²©ì, ë²”ë¡€, ë¼ë²¨)
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=1) # xì¶•
plt.axvline(0, color='black', linewidth=1) # yì¶•
plt.title("Linear Equation Graph")

# ë³´ì—¬ì¤˜!
plt.show()
```
