```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=1)
dt_clf = DecisionTreeClassifier(random_state =1)
dt_clf.fit(X_train, y_train)
pred = dt_clf.predict(X_test)
round(accuracy_score(y_test, pred), 4)
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5)

iris = load_iris()
feature = iris.data
label = iris.target

cv_accuracy = []
n_iter =0
dt_clf = DecisionTreeClassifier(random_state=1)
```

```python
feature.shape
```

```python
n_iter =0
for train_index, test_index in kfold.split(feature):
  x_train, x_test = feature[train_index], feature[test_index]
  y_train, y_test = label[train_index], label[test_index]

  dt_clf.fit(x_train, y_train)
  pred = dt_clf.predict(x_test)
  n_iter += 1

  accuracy = np.round(accuracy_score(y_test, pred), 4)
  train_size = x_train.shape[0]
  test_size = x_test.shape[0]

  print('\n#{0} 교차 검증 정확도:{1}, 학습 데이터 크기 :{2}, 검증 데이터 크기 :{3}'
    .format(n_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))
  cv_accuracy.append(accuracy)
```

```python
print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))
```

# 위처럼 결과가 나오는 경우에 일반 k-Fold 교차 검증이 아닌 계층별 교차 검증이 더 좋은 방식이다. 위와 같이 편식으로 발생하는 결과가 나오지않도록 강제로 비율을 맞춰서 골고루 먹게하는 방식이기 때문이다.
그래서 다음으로 나오는게 계층별 교차 검증(Stratified Kfold)이다.

```python
from sklearn.model_selection import StratifiedKFold

S_kfold = StratifiedKFold(n_splits=5)

cv_accuracy = []
n_iter =0

for train_index, test_index in S_kfold.split(feature, label):
  x_train, x_test = feature[train_index], feature[test_index]
  y_train, y_test = label[train_index], label[test_index]

  dt_clf.fit(x_train, y_train)
  pred = dt_clf.predict(x_test)
  n_iter += 1

  accuracy = np.round(accuracy_score(y_test, pred), 4)
  train_size = x_train.shape[0]
  test_size = x_test.shape[0]

  print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기 :{2}, 검증 데이터 크기 : {3}'
    .format(n_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스 : {1}'.format(n_iter,test_index))
  cv_accuracy.append(accuracy)

print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))
```

```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

from sklearn.model_selection import cross_val_score

iris = load_iris()
feature = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=1)

basic_scores = cross_val_score(dt_clf, feature, label)
print('\n##cross_val_score를 통한 성능 평가 :', basic_scores)
```

------------- L O O C V ----------------------

K-Fold 와 비교했을 때 데이터 학습의 일관성 유지에 좋은 방식이라고 한다. 결과적으로 모델의 평균적인 일반화 성능을 가장 꼼꼼하게 채점할 수 있다.

```python
from sklearn.model_selection import LeaveOneOut

iris = load_iris()
feature = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=1)

loocv = LeaveOneOut()
score = cross_val_score(dt_clf, feature, label, cv = loocv)

print("\n## 교차 검증 횟수 :", len(score))
print('\n## 평균 검증 정확도:', round(np.mean(score),4))
```

------------- 반복 / 그룹별 교차 검증 ----------------------

데이터 학습에서 불확실성을 최대한 줄이고자 섞고 분할해서 평가하는 사이클을 반복하는 것이 반복 교차 검증이고 그룹별 교차 검증은 데이터 안의 특정 그룹이 있다면 그 데이터를 train과 test로 나뉘게 하지않고 그룹 통째로 어느 한쪽으로 들어가게 하는 것인데 그렇게 하는 이유는 특정 그룹이 학습에도 쓰이고 시험에도 쓰이면 그 특정 그룹에 대한 과적합이 일어날 수 있기 때문에 방지하고자 함이다.

```python
from sklearn.model_selection import RepeatedStratifiedKFold

iris = load_iris()
feature = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=1)

rskfold = RepeatedStratifiedKFold()
score = cross_val_score(dt_clf, feature, label, cv = rskfold)

print("\n## 교차 검증 횟수 :", len(score))
print('\n## 평균 검증 정확도:', round(np.mean(score),4))
```
